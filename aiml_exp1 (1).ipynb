```python
import matplotlib.pyplot as plt

# Given data set
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# 3. Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(age, height, c='blue', label='Height')
plt.scatter(age, weight, c='red', label='Weight')
plt.title('Scatter Plot of Age vs. Height and Weight')
plt.xlabel('Age')
plt.ylabel('Height and Weight')
plt.legend()
plt.grid(True)
plt.show()
# 4. Bar Chart
plt.figure(figsize=(8, 6))
bar_width = 0.35
index = range(len(people))
plt.bar(index, height, width=bar_width, label='Height', color='blue')
plt.bar([i + bar_width for i in index], weight, width=bar_width, label='Weight', color='red')
plt.xlabel('People')
plt.ylabel('Height and Weight')
plt.title('Bar Chart of Height and Weight for Each Person')
plt.xticks([i + bar_width/2 for i in index], people)
plt.legend()
plt.grid(True)
plt.show()
# 5. Histogram
plt.figure(figsize=(8, 6))
plt.hist(height, bins=10, alpha=0.7, color='blue', label='Height')
plt.hist(weight, bins=10, alpha=0.7, color='red', label='Weight')
plt.title('Histogram of Height and Weight')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
